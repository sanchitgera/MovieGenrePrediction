```python
import ast
def return_genres(json):
    dictionary = ast.literal_eval(json)
    genres = []
    for key, val in dictionary.items():
         genres.append(val)
    return genres
```

```python
filtered_english_movies = english_movies[["wId", "name", "genres"]]
```

```python
filtered_english_movies["genres"].map(return_genres)
```
