## Building the Dataset

```python
import numpy 
import pandas as pd 
import matplotlib.pyplot as plt 
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.model_selection import train_test_split
```

```python
headers = ["wId", "fId", "name", "release", "revenue", "runtime", "language", "countries", "genres"]
movies = pd.read_csv("./MovieSummaries/movie.metadata.tsv", sep="\t", names= headers)
```

```python
movies.head(20)
```

```python
english_movies = movies[movies["language"] == '{"/m/02h40lc": "English Language"}']
```

```python
english_movies["wId"].count()
```

```python
english_movies.head(10)
```

```python
import ast
def return_genres(json):
    dictionary = ast.literal_eval(json)
    genres = []
    for key, val in dictionary.items():
         genres.append(val)
    return genres
```

```python
filtered_english_movies = english_movies[["wId", "name", "genres"]]
```

```python
filtered_english_movies["genres"] = filtered_english_movies["genres"].map(return_genres)
```

```python
filtered_english_movies.head(10)
```

```python
columns = ["wId", "summary"]
plot_summaries = pd.read_csv("./MovieSummaries/plot_summaries.txt", names=columns, sep="\t")
```

```python
plot_summaries.head(4)
```

```python
main = pd.merge(filtered_english_movies, plot_summaries, how="inner", on="wId")
```

```python
main.head(10)
```

```python
test = main.genres.values.flatten()
```

```python
test
```

```python
genres_counts = pd.get_dummies(main.genres.apply(pd.Series).stack()).sum(level=0)
```

```python
genres_counts.head(20)
```

```python
sums = genres_counts.sum(axis=0)
```

```python
sums.head(15)
```

```python
candidate_genres = sums.sort_values(ascending=False)[0:20]
```

```python
candidate_genres
```

```python
excluded_genres = ["Indie", "Black-and-white", "Short Film", "Action/Adventure", "Family Film", "Musical" ]
```

```python
genres = candidate_genres[~candidate_genres.index.isin(excluded_genres)]
```

```python
genres
```

```python
genres
```

```python
genres_counts = genres_counts[genres.index.values.tolist()]
```

```python
genres_counts.head(10)
```

```python
dataset = pd.merge(main, genres_counts, how="inner", left_index=True, right_index=True)
```

```python
dataset = dataset.drop(["wId", "name", "genres"], axis=1)
```

```python
dataset.head(10)
```

## Naive Baye's Classification

```python
from sklearn.pipeline import Pipeline
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import CountVectorizer
from skmultilearn.problem_transform import BinaryRelevance
from sklearn.naive_bayes import GaussianNB

text_clf = Pipeline([('vect', CountVectorizer()), ('tfidf', TfidfTransformer()), ('clf', MultinomialNB())])

classifier = BinaryRelevance(GaussianNB())

dataset = dataset.head(5000)

x = dataset['summary']

y = dataset.drop(["summary"], axis=1)

y.shape

cnt_vect = CountVectorizer(stop_words='english')
tf_transform = TfidfTransformer()
clf = MultinomialNB()

r = cnt_vect.fit_transform(x)

r = tf_transform.fit_transform(r)

x = r

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33)

model = classifier.fit(x_train, y_train)

predictions = model.predict(x_test)

from sklearn.metrics import accuracy_score

accuracy_score(y_test, predictions)

from sklearn.metrics import precision_recall_fscore_support

precision_recall_fscore_support(y_test, predictions, average='weighted')
```
